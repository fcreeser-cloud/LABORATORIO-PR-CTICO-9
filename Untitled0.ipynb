Crear un análisis simple de las ventas de un pequeño negocio, incluyendo la
creación de un DataFrame, el uso de operaciones de NumPy y la visualización de los resultados
con Matplotlib.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Crear un DataFrame de ventas
data = {
    'Producto': ['A', 'B', 'C', 'D', 'A', 'B', 'C', 'D', 'A', 'B'],
    'Ventas_Unidades': np.random.randint(10, 100, 10),
    'Precio_Unitario': np.random.uniform(5.0, 50.0, 10)
}
df_ventas = pd.DataFrame(data)

# Calcular el total de ventas por fila
df_ventas['Ventas_Totales'] = df_ventas['Ventas_Unidades'] * df_ventas['Precio_Unitario']

print("DataFrame de Ventas Inicial:")
print(df_ventas)
print("\n")

# 2. Uso de operaciones NumPy

# Calcular el total de ventas global
total_ventas_global = np.sum(df_ventas['Ventas_Totales'])
print(f"Total de Ventas Global: {total_ventas_global:.2f}")

# Calcular el promedio de ventas por unidad
promedio_ventas_unidades = np.mean(df_ventas['Ventas_Unidades'])
print(f"Promedio de Unidades Vendidas: {promedio_ventas_unidades:.2f}")

# Encontrar el producto con las ventas más altas
idx_max_ventas = np.argmax(df_ventas['Ventas_Totales'])
producto_mas_vendido = df_ventas.loc[idx_max_ventas, 'Producto']
ventas_max = df_ventas.loc[idx_max_ventas, 'Ventas_Totales']
print(f"Producto con las Ventas Totales más Altas: {producto_mas_vendido} (Ventas: {ventas_max:.2f})")

# Agrupar por producto y sumar ventas
ventas_por_producto = df_ventas.groupby('Producto')['Ventas_Totales'].sum().reset_index()
print("\nVentas Totales por Producto:")
print(ventas_por_producto)
print("\n")

# 3. Visualización de los resultados con Matplotlib

plt.figure(figsize=(10, 6))
plt.bar(ventas_por_producto['Producto'], ventas_por_producto['Ventas_Totales'], color='skyblue')
plt.xlabel('Producto')
plt.ylabel('Ventas Totales')
plt.title('Ventas Totales por Producto')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Opcional: Histograma de precios unitarios
plt.figure(figsize=(10, 6))
plt.hist(df_ventas['Precio_Unitario'], bins=5, color='lightcoral', edgecolor='black')
plt.xlabel('Precio Unitario')
plt.ylabel('Frecuencia')
plt.title('Distribución de Precios Unitarios')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Datos de calificaciones de estudiantes
data = {
    "Estudiante": ["Alice", "Bob", "Charlie", "David", "Eva", "Frank", "Grace", "Hannah", "Ian", "Jack"],
    "Matemáticas": [85, 92, 78, 90, 88, 76, 95, 89, 82, 94],
    "Ciencias": [90, 85, 80, 88, 92, 75, 95, 87, 81, 93],
    "Historia": [78, 82, 88, 90, 85, 80, 92, 84, 79, 91]
}

# Crear un DataFrame de calificaciones
df_calificaciones = pd.DataFrame(data)

print("DataFrame de Calificaciones Inicial:")
print(df_calificaciones)
print("\n")
```

```python
# 2. Uso de operaciones NumPy

# Calcular el promedio de calificaciones por asignatura
promedio_por_asignatura = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean()
print("Promedio de Calificaciones por Asignatura:")
print(promedio_por_asignatura)
print("\n")

# Calcular el promedio de calificaciones por estudiante
df_calificaciones['Promedio_General'] = df_calificaciones[['Matemáticas', 'Ciencias', 'Historia']].mean(axis=1)
print("DataFrame con Promedio General por Estudiante:")
print(df_calificaciones)
print("\n")

# Encontrar la calificación más alta y más baja en Matemáticas
calificacion_max_matematicas = np.max(df_calificaciones['Matemáticas'])
calificacion_min_matematicas = np.min(df_calificaciones['Matemáticas'])
print(f"Calificación Máxima en Matemáticas: {calificacion_max_matematicas}")
print(f"Calificación Mínima en Matemáticas: {calificacion_min_matematicas}")
print("\n")
```

```python
# 3. Visualización de los resultados con Matplotlib

# Gráfico de barras del promedio de calificaciones por asignatura
plt.figure(figsize=(10, 6))
promedio_por_asignatura.plot(kind='bar', color=['skyblue', 'lightcoral', 'lightgreen'])
plt.title('Promedio de Calificaciones por Asignatura')
plt.xlabel('Asignatura')
plt.ylabel('Promedio de Calificación')
plt.ylim(0, 100)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Gráfico de barras del promedio general por estudiante
plt.figure(figsize=(12, 7))
plt.bar(df_calificaciones['Estudiante'], df_calificaciones['Promedio_General'], color='lightgray')
plt.title('Promedio General de Calificaciones por Estudiante')
plt.xlabel('Estudiante')
plt.ylabel('Promedio General')
plt.ylim(0, 100)
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
